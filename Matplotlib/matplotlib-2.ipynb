```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
```

```python
x = np.linspace(0,5,11)
y = x **2
```

```python
x
```

```python
y
```

```python
plt.plot(x,y)


plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.title("Title")
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r-')
plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.3,0.4,0.4,0.3])
axes1.plot(x,y)
axes2.plot(y,x,'r')
axes1.set_title('Larger plot')
axes2.set_title('smaller plot')
```

```python
fig,axes = plt.subplots(nrows = 1,ncols =2,figsize =(8,2) )


axes[0].plot(x,y)
axes[1].plot(y,x)
axes[0].set_title("First_pplot")
axes[1].set_title("second plot")
```

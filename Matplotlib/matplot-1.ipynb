```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

```python
x = np.arange(0,100)
y = x*2
z = x**2
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
ax.set_xlabel("X")
ax.set_ylabel("y")
ax.set_title("Title")
```

```python
fig = plt.figure()
ax1 = fig.add_axes([0,0,1,1])
ax2 = fig.add_axes([0.2,0.4,0.3,0.3])
ax1.plot(x,y,'r')
ax2.plot(y,x)

ax1.set_xlabel("outer x")
```
